```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

metal = pd.read_csv("../dataset/metal_bands_2017.csv", encoding='unicode_escape')
metal.info()
```

```python
sns.histplot(metal["fans"],bins=50)
```

```python
bands_progressive_trash = metal[metal["style"].str.contains('Progressive') & metal["style"].str.contains('Thrash')]

fans_by_coutry = bands_progressive_trash.groupby("origin").sum().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x="fans", y="origin", data=fans_by_coutry.sort_values("fans", ascending= False))
```

```python
metal.formed = pd.to_datetime(metal.formed, errors="coerce")
sns.histplot(data=metal,x="formed")
```
