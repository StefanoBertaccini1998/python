```python
import dotenv as dot
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import os
from sqlalchemy import create_engine


stock = pd.read_csv("../dataset/stockdata.csv")
stock.describe()
```

```python
stock["Date"] = pd.to_datetime(stock["Date"],format="%Y-%m-%d")
stock.info()
```

```python
sns.lineplot(data=stock,x="Date",y="AAPL")
```

```python
stock.set_index("Date",drop=False)
stock[["MSFT","IBM","SBUX","AAPL","GSPC"]].plot(grid=True)
```

```python
stock[["MSFT","IBM","SBUX","AAPL","GSPC"]].plot()
```

```python
filter = (stock.Date.dt.year==2012 )|(stock.Date.dt.year==2013)
filter_company =["MSFT","IBM"]
loc = stock.loc[filter,filter_company]
loc.plot()
```
